# Prepare table for categorical experiment

Necessary imports:

```python
import pandas as pd
```

```python
# read experiment data
experiment = pd.read_csv('../results/categorical_exp/combined.csv')
```

```python
# preprocess dataframe
experiment = experiment.sort_values(by=["Dataset", "Degree","Algorithm"])
experiment = experiment[["Dataset", "Degree","Algorithm","Balance", "ARI_DB", "NMI_DB","Noise", 'ARI']]
```

```python
# rename columns and choose results according to ground truth clusters (2)
experiment.rename(columns = {'Degree':'k','Dataset':'Data'}, inplace = True)
experiment = experiment[experiment['k']==2]
```

```python
# prepare for latex output
table = experiment[["Data", "Algorithm","Balance","ARI_DB", "NMI_DB","Noise"]]
table = table.drop_duplicates()

print(table.to_latex(index=False,
                  float_format="{:.2f}".format,
))  
```
